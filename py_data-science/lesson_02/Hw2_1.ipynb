# Частина перша: Знайомство з Pandas.
Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання

Необхідно виконати:

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
data = pd.read_html(url,
                    match="Коефіцієнт народжуваності у регіонах України у 1950—2014 роках")
df = data[0]
```

* Вивести перші рядки таблиці за допомогою методу head

```python
df.head()
```

* Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

* Замініть у таблиці значення "—" на значення NaN

```python
df_nan = df.replace({"—": np.nan}, inplace=False)
df_nan
```

* Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df_nan.dtypes
```

* Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—". "В даній таблиці вже виконав таке завдання"

* Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df_nan.isnull().sum()
```

* Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df_nan.drop(index=df_nan.index[-1], inplace=True)
df_nan
```

* Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
mean_values = df_nan.mean()
df_nan.fillna(value=mean_values, inplace=True)
df_nan
```

```python
df_nan['1960'] = df_nan['1960'].round(2)
df_nan
```

* Отримайте список регіонів, де рівень народжуваності у 2014 році був вищим за середній по Україні

```python
mean_2014 = df_nan['2014'].mean()
mean_2014
```

```python
high_birth_rate_regions = df_nan.loc[df_nan['2014'] > mean_2014, ['регіон', '2014']]
high_birth_rate_regions.sort_values('2014', ascending=False)
```

* У якому регіоні була найвища народжуваність у 2014 році?

```python
high_bd_in_regions = df_nan.loc[df_nan['2014'] == df_nan['2014'].max(), ['регіон', '2014']]
high_bd_in_regions
```

* Побудуйте стовпчикову діаграму народжуваності по регіонах у 2014 році

```python
plt.figure(figsize = (10,6))

ax = df_nan['2014'].sort_values(ascending=False).plot(kind = 'bar', title = "Народжуваність по регіонах у 2014 році")

ax.set_xlabel('Регіони')
ax.set_ylabel('Народжуваність в тис.')
ax.set_xticklabels(labels=df_nan['регіон'], rotation=30, ha='right')

plt.show()
```

```python
df_nan.set_index('регіон', drop=True, inplace=True)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df_nan, color='r', linewidth=2)
plt.title('Динаміка народжуваності в областях за роками', fontsize=16)
plt.xlabel('Роки', fontsize=14)
plt.ylabel('Народжуваність, тис.', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid()
plt.show()
```

```python
labels = df_nan['регіон']
data = df_nan['2014']

explode = [0.10,]
explode.extend([0 for _ in range(len(data) - 1)])
plt.pie(
    data,
    labels=labels,
    shadow=False,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.title("Кругова діаграма\nпо народжуваності в 2014 р.", y=1.05)
plt.show()
```

```python
ax = df_nan.plot(kind = 'scatter',
             x = 'регіон',
             y = '2014',
             title = "Точкова візулізація", 
             figsize = (10,6))

ax.set_xlabel('Області')
ax.set_ylabel('Народжуваність тис.')
plt.xticks(rotation=45, ha='right')

plt.show()
```
